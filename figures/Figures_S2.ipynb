# Power Heatmaps — Figure S2

Each section contains two cells. The first cell computes the statistical power across the parameter grid and saves the results to a .tsv file; this computation can take several hours for large sample sizes, especially for the dominant model. We provide all pre-computed .tsv files in the `Figures_1_S1_S2/` subdirectory, so you do not need to run the first cell. Simply run the second cell to generate the plot directly from the provided saved tsv file. If you wish to recompute from scratch (e.g., to use a different type I error threshold), uncomment the first cell, run it, and the .tsv file will be overwritten before plotting.

"Rare" correspond to the zoomed in region from AF=0 to AF = 0.01

```python
import pandas as pd
_df = pd.read_csv('../data/power/power_rec_10k_1e5.tsv', sep='\t')
coord_sAA_rec10k                         = list(_df['sAA'])
coord_AF_post_rec10k                     = list(_df['AF_post'])
coord_power_rec10k                       = list(_df['power'])

import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import numpy as np

coords_Rec = pd.DataFrame({
    'sample selection': coord_sAA_rec10k,
    'AF_post':          coord_AF_post_rec10k,
    'power':            coord_power_rec10k,
})
coords_Rec = coords_Rec[coords_Rec['AF_post'] < 0.1]

fig = px.density_heatmap(coords_Rec, x='sample selection', y='AF_post', z='power',
                         nbinsx=20, nbinsy=10, histfunc='avg',
                         range_x=[1.0, 0], range_y=[0, 0.1],
                         color_continuous_scale=[[0, '#FFA500'], [0.4, '#FFE0B2'], [1, '#0091EA']])

# Grey zone: fraction wiped >= 1%  →  pA_post >= boundary curve
sAA_bnd    = np.linspace(0, 0.9999, 2000)
pA_pre_bnd = np.sqrt(0.01 / (1 - sAA_bnd))
mask       = pA_pre_bnd < 1
sAA_bnd    = sAA_bnd[mask]
pA_pre_bnd = pA_pre_bnd[mask]
pB_bnd     = 1 - pA_pre_bnd
w_bar_bnd  = sAA_bnd * pA_pre_bnd**2 + 2 * pA_pre_bnd * pB_bnd + pB_bnd**2
pA_post_bnd = (sAA_bnd * pA_pre_bnd**2 + pA_pre_bnd * pB_bnd) / w_bar_bnd
mask2       = pA_post_bnd <= 0.1

x_bnd2 = sAA_bnd[mask2]
y_bnd2 = pA_post_bnd[mask2]

# Polygon: follow boundary curve, close along top edge back to start
x_poly = list(x_bnd2) + [x_bnd2[-1], x_bnd2[0]]
y_poly = list(y_bnd2) + [0.1,         0.1       ]

fig.add_trace(go.Scatter(
    x=x_poly,
    y=y_poly,
    fill='toself',
    fillcolor='rgba(150, 150, 150, 0.6)',
    line=dict(color='grey', width=1),
    name='Fraction wiped ≥ 1%',
))

fig.update_layout(
    title='Recessive 10k — power of HWE test',
    xaxis=dict(tickfont=dict(size=20), fixedrange=False),
    yaxis=dict(tickfont=dict(size=20), fixedrange=False),
)
fig.show(config={'displayModeBar': True})
```

```python
import pandas as pd
_df = pd.read_csv('../data/power/power_dom_10k_1e5.tsv', sep='\t')
coord_sAA_dom10k                         = list(_df['sAA'])
coord_AF_post_dom10k                     = list(_df['AF_post'])
coord_power_dom10k                       = list(_df['power'])

import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import numpy as np

coords_Dom = pd.DataFrame({
    'sample selection': coord_sAA_dom10k,
    'AF_post':          coord_AF_post_dom10k,
    'power':            coord_power_dom10k,
})
coords_Dom = coords_Dom[coords_Dom['AF_post'] < 0.1]

fig = px.density_heatmap(coords_Dom, x='sample selection', y='AF_post', z='power',
                         nbinsx=20, nbinsy=10, histfunc='avg',
                         range_x=[1.0, 0], range_y=[0, 0.1],
                         color_continuous_scale=[[0, '#FFA500'], [0.4, '#FFE0B2'], [1, '#0091EA']])

# Grey zone: (1-sAA)*pA*(2-pA) >= 1%  →  small root: pA_pre = 1 - sqrt(1 - 0.01/(1-sAA))
sAA_bnd = np.linspace(0, 0.9999, 2000)
disc    = 1 - 0.01 / (1 - sAA_bnd)
mask    = disc >= 0
sAA_bnd    = sAA_bnd[mask]
pA_pre_bnd = 1 - np.sqrt(disc[mask])
pB_bnd     = 1 - pA_pre_bnd
w_bar_bnd  = sAA_bnd * pA_pre_bnd**2 + 2 * sAA_bnd * pA_pre_bnd * pB_bnd + pB_bnd**2
pA_post_bnd = (sAA_bnd * pA_pre_bnd**2 + sAA_bnd * pA_pre_bnd * pB_bnd) / w_bar_bnd
mask2       = pA_post_bnd <= 0.1

x_bnd2 = sAA_bnd[mask2]
y_bnd2 = pA_post_bnd[mask2]

x_poly = list(x_bnd2) + [x_bnd2[-1], x_bnd2[0]]
y_poly = list(y_bnd2) + [0.1,         0.1       ]

fig.add_trace(go.Scatter(
    x=x_poly,
    y=y_poly,
    fill='toself',
    fillcolor='rgba(150, 150, 150, 0.6)',
    line=dict(color='grey', width=1),
    name='Fraction wiped ≥ 1%',
))

fig.update_layout(
    title='Dominant 10k — power of HWE test',
    xaxis=dict(tickfont=dict(size=20), fixedrange=False),
    yaxis=dict(tickfont=dict(size=20), fixedrange=False),
)
fig.show(config={'displayModeBar': True})
```

```python
import pandas as pd
_df = pd.read_csv('../data/power/power_dom_500k_rare_1e5.tsv', sep='\t')
coord_sAA_dom500k_rare                   = list(_df['sAA'])
coord_AF_post_dom500k_rare               = list(_df['AF_post'])
coord_power_dom500k_rare                 = list(_df['power'])

import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import numpy as np

coords = pd.DataFrame({
    'sample selection': coord_sAA_dom500k_rare,
    'AF_post':          coord_AF_post_dom500k_rare,
    'power':            coord_power_dom500k_rare,
})
coords = coords[coords['AF_post'] < 0.01]

fig = px.density_heatmap(coords, x='sample selection', y='AF_post', z='power',
                         nbinsx=20, nbinsy=10, histfunc='avg',
                         range_x=[1.0, 0], range_y=[0, 0.01],
                         color_continuous_scale=[[0, '#FFA500'], [0.4, '#FFE0B2'], [1, '#0091EA']],
                         range_color=[0, 1])

sAA_bnd = np.linspace(0, 0.9999, 2000)
disc    = 1 - 0.01 / (1 - sAA_bnd)
mask    = disc >= 0
sAA_bnd    = sAA_bnd[mask]
pA_pre_bnd = 1 - np.sqrt(disc[mask])
pB_bnd     = 1 - pA_pre_bnd
w_bar_bnd  = sAA_bnd * pA_pre_bnd**2 + 2 * sAA_bnd * pA_pre_bnd * pB_bnd + pB_bnd**2
pA_post_bnd = (sAA_bnd * pA_pre_bnd**2 + sAA_bnd * pA_pre_bnd * pB_bnd) / w_bar_bnd
mask2       = pA_post_bnd <= 0.01

x_bnd2 = sAA_bnd[mask2]
y_bnd2 = pA_post_bnd[mask2]

x_poly = list(x_bnd2) + [x_bnd2[-1], x_bnd2[0]]
y_poly = list(y_bnd2) + [0.01,        0.01      ]

fig.add_trace(go.Scatter(
    x=x_poly,
    y=y_poly,
    fill='toself',
    fillcolor='rgba(150, 150, 150, 0.6)',
    line=dict(color='grey', width=1),
    name='Fraction wiped ≥ 1%',
))

fig.update_layout(
    title='Dominant 500k Rare — power of HWE test',
    xaxis=dict(tickfont=dict(size=20), fixedrange=False),
    yaxis=dict(tickfont=dict(size=20), fixedrange=False),
)
fig.show(config={'displayModeBar': True})
```
